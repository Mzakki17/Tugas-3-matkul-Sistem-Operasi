# **Penjelasan dari command ubuntu** #

#### 1. ls 
    Menampilkan daftar file dan direktori di direktori saat ini.
#### 2. cd
    Mengganti direktori kerja saat ini.
#### 3. pwd
    Menampilkan path direktori kerja saat ini.
#### 4. mkdir
    Membuat direktori baru.
#### 5. rmdir
    Menghapus direktori kosong.
#### 6. touch file
    Membuat file kosong atau memperbarui waktu akses file.
#### 7. nano file
    Mengedit file menggunakan editor teks Nano.
#### 8. cat file
    Menampilkan isi file.
#### 9. less file
    Menampilkan isi file satu halaman pada satu waktu.
#### 10. head file
    Menampilkan beberapa baris pertama dari file.
#### 11. tail file
    Menampilkan beberapa baris terakhir dari file.
#### 12. more file
    Menampilkan isi file secara bertahap.
#### 13. cp file1 file2
    Menyalin file1 ke file2.
#### 14. mv file1 file2
    Memindahkan atau mengganti nama file1 menjadi file2.
#### 15. rm file
    Menghapus file.
#### 16. cd /path/to/directory
    Pindah ke direktori tertentu.
#### 17. grep "text" file
    Mencari "text" di dalam file.
#### 18. find /path -name "file_name"
    Mencari file berdasarkan nama.
#### 19. locate file_name
    Mencari file dengan nama tertentu di sistem.
#### 20. echo "text"
    Mencetak teks ke terminal.
#### 21. history
    Menampilkan daftar perintah yang pernah dijalankan.
#### 22. clear
    Membersihkan layar terminal.
#### 23. date 
    Menampilkan atau mengatur tanggal dan waktu.
#### 24. cal
    Menampilkan kalender.
#### 25. df -h
    Menampilkan penggunaan disk dalam format yang mudah dibaca.
#### 26. du -sh directory
    Menampilkan ukuran direktori.
#### 27. free -h
    Menampilkan penggunaan RAM.
#### 28. uname -a
    Menampilkan informasi sistem.
#### 29. whoami 
    Menampilkan nama pengguna saat ini.
#### 30. id 
    Menampilkan ID pengguna dan grup.
#### 31. ps 
    Menampilkan proses yang berjalan saat ini.
#### 32. ps aux 
    Menampilkan semua proses yang berjalan di sistem.
#### 33. kill PID  
    Menghentikan proses berdasarkan PID.
#### 34. killall process_name 
    Menghentikan semua proses dengan nama tertentu.
#### 35. bg 
    Menjalankan proses di background.
#### 36. fg
    Membawa proses background ke foreground.
#### 37. wget URL 
    Mengunduh file dari URL.
#### 38. curl URL 
    Mengambil atau mengirim data menggunakan URL.
#### 39. ping address 
    Mengecek konektivitas jaringan ke alamat tertentu.
#### 40. ifconfig 
    Menampilkan informasi jaringan
#### 41. ip a 
    Menampilkan informasi antarmuka jaringan.
#### 42. ssh user@hostname 
    Mengakses remote server menggunakan SSH.
#### 43. scp file user@hostname:/path/to/destination 
    Menyalin file ke remote server menggunakan SSH. 
#### 44.alias ll='ls -al' 
    Membuat alias untuk perintah. 
#### 45. export VAR=value 
    Menetapkan variabel lingkungan.
#### 46. uname -r 
    Menampilkan versi kernel Linux.
#### 47. uptime 
    Menampilkan berapa lama sistem telah berjalan.
#### 48. who 
    Menampilkan siapa saja yang sedang login ke sistem.
#### 49. env 
    Menampilkan semua variabel lingkungan.
#### 50. sudo 
    Menjalankan command dengan hak akses superuser.
